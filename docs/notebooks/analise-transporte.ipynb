```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir o domínio (de -10 a 10, com 100 pontos)
x = np.linspace(-10, 10, 100)

# 2. Definir a função
y = x**2  # Exemplo: f(x) = x^2

# 3. Plotar
plt.plot(x, y, label="f(x) = x²")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Suponha que y são os dados de posição do ônibus
dy = np.gradient(y, x) # Calcula a variação (velocidade)

plt.plot(x, y, label="Posição (f)")
plt.plot(x, dy, label="Velocidade (f')") # A derivada!
```

```python
# %pip install sympy
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir a variável simbólica
x_sym = sp.symbols('x')

# 2. Definir a função simbólica
f_sym = x_sym**3 - 3*x_sym**2 + 2

# 3. Calcular a Derivada (Mágica acontece aqui)
f_prime_sym = sp.diff(f_sym, x_sym)
print(f"A derivada é: {f_prime_sym}") # Vai imprimir: 3*x**2 - 6*x

# 4. Transformar em função Python para plotar (Lambdify)
f_numeric = sp.lambdify(x_sym, f_sym, 'numpy')
f_prime_numeric = sp.lambdify(x_sym, f_prime_sym, 'numpy')

# 5. Gerar dados e plotar
x_vals = np.linspace(-2, 4, 100)
plt.plot(x_vals, f_numeric(x_vals), label=f"f(x) = {f_sym}")
plt.plot(x_vals, f_prime_numeric(x_vals), label=f"f'(x) = {f_prime_sym}", linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

```python
# Exemplo de Limite com furo
x = np.linspace(-5, 5, 100)
y = np.sin(x) / x  # O Python vai dar erro ou aviso em x=0

plt.plot(x, y)

# Adiciona a "bolinha vazia" no limite (x=0, y=1)
plt.plot(0, 1, marker='o', markersize=10, markerfacecolor='white', markeredgecolor='red')
plt.annotate('O Limite é aqui', xy=(0, 1), xytext=(2, 0.8), arrowprops=dict(arrowstyle='->'))

plt.show()
```
